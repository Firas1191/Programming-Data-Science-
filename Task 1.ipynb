### Requirement FR1.1 - Find the arithmetic mean

```python
# This Code is to call the Data from the Data List. 
List = [i.strip() for i in open 
        ("/Users/macbookpro/Desktop/Data Science/Programming /Assessment/task1.dat").readlines()]
List_of_Data = (list(map(int, List)))
# the Equation to get the mean. 
count = len(List_of_Data)
mean = sum(List_of_Data) / count

# This code is used to have the result as 1 decomel Place 
x= mean
float = x
format_float1 = "{:.2f}".format(float) # to get the results in 2dp 

#Printing out the results. 
print("The mean is qual to =", format_float1)
```

### Requirement FR1.2 - Find the standard deviation 

```python
# getting the right equation of Stander Diviation
variance = sum([((x - mean) ** 2) for x in List_of_Data]) / count
std = variance ** 0.5

# This code is used to have the result as 1 decomel Place 
float = std
format_float2 = "{:.2f}".format(float) # to get the results in 2dp 

#printing the Results 
print("The Standard Deviation is qual to =",format_float2)
```
